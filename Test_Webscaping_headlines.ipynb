```python
driver= webdriver.Chrome(executable_path="C:/Users/egust/python/chromedriver.exe") # Driver location on your computer
webdriver.remote.webelement.WebElement

url = 'https://coinmarketcap.com/currencies/ethereum/news/' # website to scrape
driver.get(url) # Scraping
time.sleep(5)
```

```python
websites = driver.find_elements_by_xpath("//a[@class='svowul-0 jMBbOf cmc-link']")
```

```python
websites_ = []

for i in range(len(websites)):
    websites_.append(websites[i].get_attribute('href'))

websites_
```

```python
column = {'website':websites_}
df_website = pd.DataFrame(data=column)
df_website.head()
```

```python
job_list = driver.find_elements_by_xpath("//div[@class='svowul-2 hMwffQ']")
job_list
```

```python
headline = []
news_center = []
time = []

for element in job_list:
    headline.append(element.text.split('\n')[0])
    news_center.append(element.text.split('\n')[1])
    time.append(element.text.split('\n')[2])

print(len(headline))
print(len(news_center))
print(len(time))
```

```python
columns= {'headlines':headline, 'news_center':news_center, 'time':time}
df_ether_headlines = pd.DataFrame(data=columns)
df_ether_headlines
```

```python
# Join the dataframes to include the websites for each article
df_ether_headlines = df_ether_headlines.join(df_website)
df_ether_headlines.head()
```

```python
for i in range(0,len(df_ether_headlines)):
    # Create a driver to scrape news sites 
    driver2= webdriver.Chrome(executable_path="C:/Users/egust/python/chromedriver.exe") # Driver location on your computer
    webdriver.remote.webelement.WebElement
    
    # Using the index from index finder funtion put specific url to grab the correct date
    url= df_ether_headlines.website[i]
    driver2.get(url) # Scraping
    time.sleep(5)
    
    if df_ether_headlines.date_ef[i] == 'Needs personally add in':
        df_ether_headlines['date'][i] = df_ether_headlines['date'][i]
    else:
        date = driver2.find_elements_by_xpath(df_ether_headlines.date_ef[i])
        df_ether_headlines['date'][i] = date[0].text
```
