```python
table = driver.find_elements_by_xpath("//div[@class='sc-16r8icm-0 elzRBB container']")
```

```python
# Put the table text in a list 
table_text = table[0].text
# Need to clean data
table_hd = table_text.split('\n')
table_hd
```

```python
# Change Market Cap into Market_Cap to keep it from splitting 
table_hd[2] = table_hd[2].replace('Market Cap','Market_Cap')
table_hd[2]
```

```python
table_hd_clean = [i.split(' ') for i in table_hd[3:366]]
table_hd_clean[0]
```

```python
table_hd_clean[0][0]+' '+table_hd_clean[0][1]+table_hd_clean[0][2]
```

```python
Date = []
Open = []
High = []
Low = []
Close = []
Volume = []
Market_Cap = []

for i in range(len(table_hd_clean)):
    Date.append(table_hd_clean[i][0]+' '+table_hd_clean[i][1]+table_hd_clean[i][2])
    Open.append(table_hd_clean[i][3])
    High.append(table_hd_clean[i][4])
    Low.append(table_hd_clean[i][5])
    Close.append(table_hd_clean[i][6])
    Volume.append(table_hd_clean[i][7])
    Market_Cap.append(table_hd_clean[i][8])
```

```python
data = {'date':Date, 'open':Open,
        'high':High, 'low':Low,
        'close':Close, 'volume':Volume,
       'market_cap':Market_Cap}
```

```python
historical_data = pd.DataFrame(data)
historical_data
```

```python
historical_data.date[0]
```

```python
# convert the 'Date' column to datetime format
from datetime import datetime
datetime.strptime(historical_data.date[0], '%b %d,%Y')
```

```python
for i in range(len(historical_data)):
    historical_data.date[i] = datetime.strptime(historical_data.date[i], '%b %d,%Y')
```

```python
# Convert the date column from string to datetime
historical_data.date = pd.to_datetime(historical_data.date).dt.date
```
